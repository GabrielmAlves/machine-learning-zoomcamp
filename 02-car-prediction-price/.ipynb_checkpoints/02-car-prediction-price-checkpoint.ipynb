```python
## now we need to split our dataset into three different parts:
## train, validation and test
## we train the model using the train set, validate it using the validation set and
## use the test set on the end to see if the model is actually good and not just predicting right by chance

## a suggested % of the division of the sets is 60, 20, 20, respectively.
## but we need to know how much 20% represents out of the total number of samples in the dataset

dataset_size = len(df) ## total size of the dataset

validation_set_size = int(len(df) * 0.2) ## 20%
test_set_size = int(len(df) * 0.2)
train_set_size = dataset_size - validation_set_size - test_set_size
```

```python
## linear regression formula

## g(xi) = w0 + sum of wi * xij

## g -> model
## xi -> a car (row of feature matrix)
## w0 -> bias
## wi -> the weight given for each element of the feature vector
## xij -> an element of the feature vector
```

```python
xi = [453, 11, 86] ## feature vector
w0 = 7.17 ## bias -> what is predicted about the car if we don't know anything about it
w = [0.01, 0.04, 0.002] ## feature weights

## 7.17 + 453*0.01 + 11*0.04 + 86*0.002
```

```python
## for every element of feature vector, we multiply it for corresponding feature weight
## and everything goes to prediction
def linear_regression(xi):
    n = len(xi)

    prediction = w0

    for element in range(n):
        prediction = prediction + w[element] * xi[element]
    
    return prediction
```

```python
linear_regression(xi)
```

```python
np.expm1(linear_regression(xi))
```

```python
## linear regression formula

## g(xi) = w0 + sum of wi * xij

## g -> model
## xi -> a car (row of feature matrix)
## w0 -> bias
## wi -> the weight given for each element of the feature vector
## xij -> an element of the feature vector

## the sum part of g(xi) = w0 + sum of wi * xij is a dot product
## so this can be rewritten as w0 + dot(xi transposed * weight vector) because this is a vector x vector multiplication
w_new = [w0] + w
def dot(xi, w):
    n = len(xi)

    result = 0.0

    for j in range(n):
        result = result + xi[j] * w[j]

    return result

def linear_regression(xi):
    xi = [1] + xi
    return dot(xi, w_new)
```

```python
## since the w0 bias is alone, we can try to bring together to the dot product
## we can imagine that there is an extra feature on the car matrix and this feature is always 1
## if this is 1, it means that it doesn't affect it, it doesn't exist

## the vector w of weights become -> [w0, w1, w2, ..., wn] n+1 dimension
## xi vector = [1, xi1, xi2, ..., xin] we added the one on the front
## so when we the dot product between them, w*xi, we multiply w0 by 1, which leaves us with w0 itself
## w*xi = w0* 1 + w1 * xi1 +  ... -> w0 + w1 * x1 + w2 * x2 + ... = equivalent to what we had
## it means we can use the entire dot product notation for the entire linear regression
```

```python
linear_regression(xi)
```

```python
## we need to implement matrix x vector multiplication
## x1 [1, x11, x12, x13, ..]     w [w0, w1, w2, w3, ...] = [x1*w, x2*w, x3*w, ...] -> predictions for each car (each x)
## x2 [1, x11, x12, x13, ..]
## x3 [1, x11, x12, x13, ..]
## x4 [1, x11, x12, x13, ..]
```

```python
xi = [453, 11, 86] ## feature vector
w0 = 7.17 ## bias -> what is predicted about the car if we don't know anything about it
w = [0.01, 0.04, 0.002] ## feature weights
w_new = [w0] + w

x1 = [1, 148, 24, 1385]
x2 = [1, 132, 25, 2031]
x3 = [1, 453, 11, 86]

X = [x1, x2, x3]
X = np.array(X)
X
```

```python
def linear_regression(X):
    return X.dot(w_new)

linear_regression(X) ## predictions
```

```python
## how do we come with the weights values?
## we always want that the predicted value to be close to y
## ideally it should be equal to y (often not possible)
## we need to find a way to find w (weights)
## if X has an inverse, multiply it on both sides
## Xinverted * X * w = Xinverted * y
## crosses Xinverted and X = obtain identity matrix
## I * w = Xinverted * y -> w = Xinverted * y
## X is often a rectangular matrix, m x m+1 dim and it is not square, so the inverse doesn't exist
## the solution above would not be acceptable
## but there is an approximate solution by multiplying for the transposed matrix
## Xt * X * W = Xt * y ------- Xt * X is an gram matrix and for this matrix the inverse exist, since it is always squared
## gram matrix is a m+1 x m+1 dim
## so we multiply everything by the inverse of the gram matrix
## (Xt *X)inverted * Xt * X * w = (Xt *X)inverted * Xt * y
## so we cross Xt *X)inverted * Xt * X, which gives us an identity matrix
## and we are left with w = (Xt *X)inverted * Xt * y
## it is the closest possible solution for w values
```

```python
X = [
    [148, 24, 1385],
    [132, 25, 2031],
    [453, 11, 86],
    [158, 24, 185],
    [172, 25, 201],
    [413, 11, 86],
    [38, 54, 185],
    [142, 25, 431],
    [453, 31, 86],
]
X = np.array(X)
X
```

```python
y = [10000, 20000, 15000, 20050, 10000, 20000, 15000, 25000, 12000]
```

```python
## we are implementing w = (Xt * X)inversed * Xt * y
XTX = X.T.dot(X) ## gram matrix
```

```python
XTX_inv = np.linalg.inv(XTX) ## gram matrix inversed
```

```python
w_full = XTX_inv.dot(X.T).dot(y)
w0 = w_full[0]
w = w_full[1:]

w0, w
```

```python
def train_linear_regression(X, y):
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])

    XTX = X.T.dot(X) ## gram matrix
    XTX_inv = np.linalg.inv(XTX) ## gram matrix inversed
    w_full = XTX_inv.dot(X.T).dot(y)

    return w_full[0], w_full[1:]
```

```python
train_linear_regression(X, y)
```

```python
## evaluating regression models
## RMSE -> Root Mean Squared Error

## g(xi) produces a prediction for xi and we have the actual value yi for this observation i
## so we take the difference between them and square it (g(xi) - yi)2
## and we take the average of these differences
## so we have a sum that goes from one to m (m is the number of observations) and we take the average
## and then the square root
## square root of sum(g(xi) - yi)2 divided by m
## g(xi) = prediction
## yi = actual price

## example:

## y_pred =  [10, 9, 11, ..., 10]
## y_train = [9, 9, 10.5, ..., 11.5]

## differences = [1, 0, 0.5, ..., -1.5]
## squared error = [1, 0, 0.25, ..., 2.25]
## average = 1 + 0 + 0.25 + 2.25 and divide it by 4 (in this case) = 0.875
## mean squared error = 0.875
## and then we take the square root of 0.875 = 0.93
```

```python
## implementing RMSE
## formula -> root square of the average of the sum of the differences between predictions and
## target variable

def rmse(y, y_pred):
    squared_error = (y - y_pred) ** 2
    mse = squared_error.mean()

    return np.sqrt(mse)
```

```python
rmse(y_train, y_pred)
```

```python
base = ['engine_hp', 'engine_cylinders', 'highway_mpg', 
        'city_mpg', 'popularity']

def prepare_X(df):
    df_num = df[base]
    df_num = df_num.fillna(0)
    X = df_num.values
    return X
```

```python
rmse(y_validation, y_pred)
```

```python
## feature engineering
## created the feature age which tell us the age of each car
## 2017 - df.year is because 2017 is when the data was created
def prepare_X(df):
    df = df.copy()
    
    df['age'] = 2017 - df.year
    features = base + ['age']
    df_num = df[features]
    df_num = df_num.fillna(0)
    X = df_num.values
    return X
```

```python
X_train = prepare_X(df_train)
w0, w = train_linear_regression(X_train, y_train)
## training part


## validation part
X_val = prepare_X(df_validation)
y_pred = w0 + X_val.dot(w)

rmse(y_validation, y_pred)
```
