```python
import pandas as pd
import numpy as np

import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
!head CreditScoring.csv
```

```python
df = pd.read_csv(data)
```

```python
df.columns = df.columns.str.lower()
```

```python
df.head()
```

```python
status_values = {
    1: 'ok',
    2: 'default',
    0: 'unknown'
}

df.status = df.status.map(status_values)
```

```python
df.head()
```

```python
home_values = {
    1: 'rent',
    2: 'owner',
    3: 'private',
    4: 'ignore',
    5: 'parents',
    6: 'other',
    0: 'unknown'
}

df.home = df.home.map(home_values)

marital_values = {
    1: 'single',
    2: 'married',
    3: 'widow',
    4: 'separated',
    5: 'divorced',
    0: 'unknown'
}

df.marital = df.marital.map(marital_values)

records_values = {
    1: 'no',
    2: 'yes',
    0: 'unknown'
}

df.records = df.records.map(records_values)

job_values = {
    1: 'fixed',
    2: 'partime',
    3: 'freelance',
    4: 'others',
    0: 'unknown'
}

df.job = df.job.map(job_values)
```

```python
df.head()
```

```python
df.describe()
```

```python
for column in ['income', 'assets', 'debt']:
    df[column] = df[column].replace(to_replace=99999999, value=np.nan)
```

```python
df.describe().round()
```

```python
df.status.value_counts()
```

```python
df = df[df.status != 'unknown'].reset_index(drop=True)
```

```python
from sklearn.model_selection import train_test_split

df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=11)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=11)
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
```

```python
y_train = (df_train.status == 'default').astype('int').values
y_val = (df_val.status == 'default').astype('int').values
y_test = (df_test.status == 'default').astype('int').values
```

```python
del df_train['status']
del df_val['status']
del df_test['status']
```

```python
df_train
```

```python
def assess_risk(client):
    if client['records'] == 'yes':
        if client['job'] == 'partime':
            return 'default'
        else:
            return 'ok'
    else:
        if client['assets'] > 6000:
            return 'ok'
        else:
            return 'default'
```

```python
xi = df_train.iloc[0].to_dict()
```

```python
assess_risk(xi)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_extraction import DictVectorizer
from sklearn.metrics import roc_auc_score
```

```python
train_dicts = df_train.to_dict(orient='records')
dv = DictVectorizer(sparse=False)
```

```python
X_train = dv.fit_transform(train_dicts)
```

```python
dt = DecisionTreeClassifier(max_depth=2)
dt.fit(X_train, y_train)
```

```python
val_dicts = df_val.to_dict(orient='records')
X_val = dv.transform(val_dicts)
```

```python
y_pred = dt.predict_proba(X_train)[:, 1]
auc = roc_auc_score(y_train, y_pred) ## overfitting
print(f"train auc {auc}")

y_pred = dt.predict_proba(X_val)[:, 1]
auc = roc_auc_score(y_val, y_pred)
print(f"validation auc: {auc}")
```

```python
from sklearn.tree import export_text
```

```python
print(export_text(dt, feature_names=dv.get_feature_names_out()))
```

```python
data = [
    [8000, 3000, 'default'],
    [2000, 1000,'default'],
    [0, 1000,'default'],
    [5000, 1000, 'ok'],
    [5000, 1000, 'ok'],
    [4000, 1000, 'ok'],
    [9000, 500, 'ok',],
    [3000, 2000, 'default'],
] ## assets, status

df_example = pd.DataFrame(data, columns=['assets', 'debt', 'status'])
df_example
```

```python
df_example.sort_values('assets')
```

```python
Ts = [0, 2000, 3000, 4000, 5000, 8000] ## cut the dataset into left and right datasets according to the value of T (threshold)
```

```python
from IPython.display import display

for T in Ts:
    print(f"Threshold: {T}")
    df_left = df_example[df_example.assets <= T]
    df_right = df_example[df_example.assets > T]

    display(df_left)
    print(df_left.status.value_counts(normalize=True))
    display(df_right)
    print(df_right.status.value_counts(normalize=True))
```

```python
## verify which split is the best
## predict misclassification rate: when we predict everyone as default, for example, how many errors did we made?
```

```python
df_example.sort_values('debt')
```

```python
thresholds = {
    'assets': [0, 2000, 3000, 4000, 5000, 8000],
    'debt': [500, 1000, 2000]
}

for feature, Ts in thresholds.items():
    print('##########')
    print(feature)
    for T in Ts:
        print(f"Threshold: {T}")
        df_left = df_example[df_example[feature] <= T]
        df_right = df_example[df_example[feature] > T]
    
        display(df_left)
        print(df_left.status.value_counts(normalize=True))
        display(df_right)
        print(df_right.status.value_counts(normalize=True))

        print()

    print('#################')
```
